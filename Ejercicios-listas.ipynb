## Ejercicio 1

```python
cant_products = int(input('Digite la cantidad de productos a procesar:'))
prices = []
products = []
stock_sales = []
ingresos = []

for i in range(cant_products):
    product_name = input('Ingrese el nombre del producto: ')
    price_product = float(input(f'Producto {i+1} - Ingrese el precio de {product_name}: '))
    stock_sale = int(input(f'Producto {i+1} - Digite la cantidad vendida de {product_name}: '))

    products.append(product_name)
    prices.append(price_product)
    stock_sales.append(stock_sale)

    ingreso = prices[i] * stock_sales[i]
    ingresos.append(ingreso)


print(f'Ingresos por producto: {ingresos}')
print(f'Ingreso total: $ {sum(ingresos)}')
```

## Ejercicio 2

```python
productos_destacados = []
for ingreso in ingresos :
    if ingreso > 10000:
        productos_destacados.append(ingreso)
print(f'Lista de ingresos: {ingresos}')

if len(productos_destacados) == 0:
    print('No hay ningun ingreso destacado')
else:
    print(f'Productos destacados (ingresos > $10,000): {productos_destacados}')
```

## Ejercicio 3

```python
cantidad_procesar = int(input('¿Cuántos productos desea ingresar?: '))
productos = []
ingreso_productos = []

for i in range(cantidad_procesar):
    product_name = input(f'Producto {i+1} - Nombre: ') 
    ingreso_prod = float(input(f'Producto {i+1} - Ingreso: $'))

    productos.append(product_name)
    ingreso_productos.append(ingreso_prod)

mayor_ingreso = max(ingreso_productos)
product_win = ingreso_productos.index(mayor_ingreso)
print(f'El producto con mayor ingreso es: {productos[product_win]} con un ingreso de: ${mayor_ingreso}')
```
