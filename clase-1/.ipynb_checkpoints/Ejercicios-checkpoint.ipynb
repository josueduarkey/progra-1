## 1. Cálculo de impuesto

```python
unidades = input('Digite las unidades consumidas: ')

while not unidades.isdigit():
    unidades = input('Digite la unidad en enteros: ')

unidades = int(unidades)

if unidades >= 101 and unidades <= 200 :
    impuesto = float(unidades * 0.5)
elif unidades >= 201 :
    impuesto = float(unidades * 0.75)
else :
    impuesto = 0


print(f'El impuesto aplicado es de ${impuesto}')
```

# 2. El número mágico

```python
number = input('Digite un número cualquiera: ')
while not number.isdigit():
    number = input('Digite un número: ')

number = int(number)
if (number % 7 == 0) and not (number % 5 == 0):
    print('su número es mágico')
else 
    print('su número NO es mágico')
```

## 3. Elevador lógico

```python
def is_digit(number, message) :
    while not number.isdigit():
        number = input(message)
    return int(number)

piso = input('Digite el piso que se encuentra: ')
elevator_a = input('Digite el piso que se encuentra el elevador - A: ')
elevator_b = input('Digite el piso que se encuentra el elevador - B: ')

message_piso = 'Digite el piso que se encuentra: '
message_elevator_a = 'Digite el piso que se encuentra A: '
message_elevator_b = 'Digite el piso que se encuentra B: '

piso = is_digit(piso, message_piso)
elevator_a = is_digit(elevator_a, message_elevator_a)
elevator_b = is_digit(elevator_b, message_elevator_b)

distancia_a = abs(piso - elevator_a)
distancia_b = abs(piso - elevator_b)

if distancia_a == distancia_b :
    print('ELEVADOR A')
elif distancia_a < distancia_b:
    print('ELEVADOR A')
else :
    print('ELEVADOR B')
```
