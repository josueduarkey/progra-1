```python
# Crear un programa que pida al usuario su edad y su DUI.
# Si el usuario tiene 18 años o más y el DUI tiene 9 digitos
# Mostrar el mensaje: "Datos Correctos, el ciudadano puede votar"
# Sino: Error, Verificar los datos

age = int(input('Digite su edad: '))
DUI = input('Digite su DUI: ')



if age >= 18 :
    if len(DUI) == 9:
        print("Datos correctos")
    else:
       print("DUI incorrecto")
else:
    print('El ciudadano no puede votar')
```
