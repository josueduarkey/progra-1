```python
# Imprimir los numeros del 1 al 10
print(1)
print(2)
print(3)
print(4)
print(5)
print(6)
print(7)
print(8)
print(9)
print(10)
```

```python
# Introducción al bloque "for"
# for "variable" in "lista/rango":
#    intruction1
#    instrution2
#    ...
```

```python
# Ejemplo: imprimir los números del 1 a 10 usando bloque "for"
for i in range(10):
    print(i)
```

```python
# Controlando el valor de inicio:
for i in range(1, 10):
    print(i)
```

```python
# Controlando el paso (por defecto 1):
for i in range(1, 10, 2):
    print(i)
```

```python
# Controlando el paso (por defecto 1):
for i in range(2, 10, 2):
    print(i)
```

```python
# Controlando el paso (por defecto 1):
for i in range(30, 10, -2):
    print(i)
```

```python
# Ejemplo practico
import math
def operation(a,b):
    r2 = (a**2)+(b**2)          # a^2 + b^2
    r = math.sqrt(r2)
    return r
```

```python
A = int(input('Ingrese lado A: '))
B = int(input('Ingrese lado B: '))
r = operation (A,B)
print(f'La hipotenusa de los lados a={A} y b={B} es: {round(r,2)}')
```

```python
for i in range(1,6):
    print(f'Solicitado lados de triangulo {i}/5')
    A = int(input('Ingrese lado A: '))
    B = int(input('Ingrese lado B: '))
    r = operation (A,B)
    print(f'La hipotenusa de los lados a={A} y b={B} es: {round(r,2)}')
```
