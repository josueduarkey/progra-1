## 2.6 Principio de Iteración

```python
from random import randint
cant_students = input('Digite la cantidad de estudiantes a ingresar en esta materia:')

while not cant_students.isdigit() :
    cant_students = input('digite un número: ')

cant_students = int(cant_students)
for i in range(1, cant_students) :
    id = randint(1, 1000)
    name = input('Digite el nombre: ')
    last_name = input('Digite el apellido: ')
    carnet = input('Digite su numero de carnet: ')

    print(f'Estudiante {id}: {name} {last_name} - {carnet}')
    
    
```
