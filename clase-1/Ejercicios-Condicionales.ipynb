### 1. Calculadora de IMC (Índice de Masa Corporal):

Reto: Pide peso y altura. Calcula el IMC y clasifica si la persona tiene peso bajo, normal o sobrepeso usando if-elif-else.

```python
peso = float(input('Digite su peso en Kg: '))
altura = float(input('Digite su altura en metros: '))

imc = round(peso/altura**2, 2)

if imc < 16.5 :
    print(f' Tu IMC es {imc}, Bajo peso severo')
elif imc < 18.5 :
    print(f' Tu IMC es {imc}, Bajo peso')
elif imc < 24.9 and imc > 18.5 :
    print(f' Tu IMC es {imc}, Normal peso')
elif imc < 29.9 and imc > 25 :
    print(f' Tu IMC es {imc}, Sobre peso')
elif imc > 30 :
    print(f' Tu IMC es {imc}, Anda al gym')
else :
    print(f'Digite sólo valores válidos')
    
```

### 2. El Juego de Adivinar el Número:

Reto: Genera un número aleatorio con import random. El usuario tiene 5 intentos. El programa debe decir "Más alto" o "Más bajo" después de cada intento.

```python
from random import randint;

real_number = randint(1, 10)
user_win = False
print(real_number)
for i in range(5) :
    user_number = int(input('Digite el número que crees que sea'))
    if user_number < real_number :
        print('Más Alto')
        continue
    elif user_number > real_number : 
        print('Más Bajo')
        continue
    else :
        print('FELICIDADES ACERTASTE')
        user_win = True
        break

if not user_win :
    print(f'Lo siento, te quedaste sin intentos. El número era {real_number}.')
```

### 3. Validador de Contraseñas:
Reto: Crea una función que reciba un string y verifique si tiene más de 8 caracteres, al menos un número y una mayúscula. Retorna True o False.

```python
def password_is_valid(password) :
    exist_number = False
    is_upper = False
    if len(password) < 8: 
        print('Digite una contraseña con más de 8 caracteres')
        return
    for char in password :
        if char.isdigit() :
            exist_number = True
        if char.isupper() :
            is_upper = True
            
    if exist_number and is_upper :
        return True

    return False
    
password_is_valid('HolaMundo')
```

### 4. Gestor de Tareas (To-Do List):

Reto: Crea un programa con un menú (1. Agregar, 2. Ver, 3. Eliminar, 4. Salir). Las tareas deben guardarse en un archivo .txt o .json para que no se borren al cerrar el programa.


```python

import sys, os
# Menu of tasks  
def menu(option, user_tasks) :
    print('----------------------')
    print('Que deseas realizar: ')
    print('1. Agregar una tarea')
    print('2. Ver todas tus tareas')
    print('3. Eliminar una tarea')
    print('4. Salir')
    print('----------------------')
    message = 'Digite la opción nuevamente (1-4): '
    option = is_number(option, message)
    
    match option:
        case 1: 
            new_task = input('Digite su tarea: ')
            add_task(new_task, user_tasks)
        case 2: print(watch_tasks(user_tasks))
        case 3: 
            index_delete = input('Que tarea desea eliminar: ')
            message = 'Que tarea desea eliminar:'
            index_delete = is_number(index_delete, message)
            
            delete_task(index_delete, user_tasks)
        case 4: exit_menu()
        case _: return "Digite un valor del 1 al 4"

def clear_console():
    os.system('cls' if os.name == 'nt' else 'clear')

def save_tasks(tasks) :
    list_of_taks = watch_tasks(tasks)
    with open("./todolist.txt", "w") as file:
        file.writelines(list_of_taks)

# Adding Task
def add_task(task_name, task_list) :
    while not task_name.strip() :
        task_name = input('Digite la tarea nuevamente: ')
        
    all_tasks = len(task_list)
    task_list.append(task_name)
    print(f'El usuario agrego la tarea {all_tasks + 1}. {task_name}')
    save_tasks(task_list)
    
# List tasks
def watch_tasks(all_tasks) :
    list_text = ''
    if not all_tasks:
        return 'No hay tareas'
        
    for i, task in enumerate(all_tasks) :
        list_text += f'{i + 1}. {task} \n'
    return list_text
    
# Delete one Task
def delete_task(index, task_list) :
    if not len(task_list) :
        print('No existe ninguna tarea aún')
        return
    if not (index >= 1 and index <= len(task_list)) :
        print('No existe, dame otro')
        return 
    task_list.pop(index - 1)
    

# Exit
def exit_menu() :
    print('Muchas gracias por tu visita')
    sys.exit() 

def is_number(num, message) :
    while not num.isdigit() :
        num = input(message)
    return int(num)

user_tasks = []
while True :
    user_option = input('Seleccione la opción (1-4): ')
    menu(user_option, user_tasks)
```
