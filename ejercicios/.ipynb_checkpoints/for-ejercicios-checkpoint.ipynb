### Tarea 1. 
Crear un programa en Python que solicite al usuario 5 precios de productos. A continuación, el
programa mostrará el promedio de los precios ingresados.

```python
sum = 0
for i in range(5):
    price = float(input('Digite el precio del producto'))
    sum += price
average = sum / 5
print(f'El promedio final es de ${average}')
```

### Tarea 2. 
Crear un programa que solicite al usuario una cadena de texto que puede ser una palabra o frase.
A continuación, el programa contará la cantidad de vocales disponibles en la cadena de texto ingresada.
Finalmente, se mostrará en pantalla la suma resultante.

```python
text = input('Digite una palabra o frase: ')
sum_vowels = 0
for i in text:
    if i in "aeiouAEIOU": 
        sum_vowels += 1
print(sum_vowels)
```

### Tarea 3. 
Solicita al usuario una cadena de texto que puede ser una palabra o frase, luego, el programa
mostrará a la cadena de texto ingresada escrita al revés. La restricción es que esta tarea se debe realizar
utilizando un bloque for.
                     


```python
text = input('Digite una palabra o frase')
text_reves = ''

for char in text:
    text_reves = char + text_reves
print(text_reves)
```

### Tarea 4. 
Crear un programa que solicita al usuario dos números enteros y positivos. Posteriormente, el
programa muestra los números pares del rango ingresado.

```python
num1 = int(input('Digite el primer número: '))
num2 = int(input('Digite el segundo número: '))

for i in range(num1, num2, ):
    if i % 2 == 0:
        print(i)
```

### Tarea 7. 
Pide un número al usuario y suma sus dígitos hasta que quede un solo dígito. Ejemplo: Si el
usuario ingresa 9875, entonces: 9+8+7+5 = 29, luego: 2+9 = 11, luego: 1+1 = 2.

```python
number = input('Digite un número: ')
sum = 0
for i in number:
    num_int = int(i)
    sum += num_int
    
print(sum)
```
