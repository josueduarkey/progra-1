### Tarea 1. 
Crear un programa en Python que solicite al usuario dos números enteros y muestre en pantalla:
cuál es el mayor, cuál es el menor, o si son iguales. Sugerencia: investiga como usar la estructura if-elif-
else

```python
number_one = int(input('Digite el primer número: '))
number_two = int(input('Digite el segundo número: '))

if number_one > number_two:
    print(f'El número {number_one} es mayor que {number_two}')
elif number_one < number_two :
     print(f'El número {number_two} es mayor que {number_one}')
else :
    print('Los números son iguales')
```

### Tarea 2.
Solicita al usuario una edad (entero). El programa debe indicar: si es menor de edad, si es mayor
de edad, y si aplica, si es adulto mayor (por ejemplo, 60+). Sugerencia: usa condiciones con rangos.

```python
age = int(input('Digite su edad: '))

if age < 0 :
    print('Error de capa 8') 
elif age < 18 :
    print('Eres menor de edad')
elif age < 60 :
    print('Eres mayor de edad')
else :
    print('Eres mayor de edad y adulto mayor')
    
```

### Tarea 3. 
Pide al usuario una calificación de 0 a 100. El programa debe imprimir:
o "Excelente" (90–100)
o "Bueno" (80–89)
o "Suficiente" (70–79)
o "Insuficiente" (<70)
• Si el usuario ingresa un valor fuera de 0–100, mostrar "Valor inválido".

```python
grade = int(input('Ingrese la nota del estudiante'))

if grade < 0 or grade > 100 :
    print('Error de capa 8') 
elif grade < 70 :
    print('Insuficiente')
elif grade >= 70 and grade <= 79 :
    print('Bueno')
elif grade >= 80 and grade <= 89 :
    print('Suficiente')
else :
    print('Excelente')
```

### Tarea 4. 
Solicita al usuario una palabra. El programa debe determinar si: la palabra está en mayúsculas, en
minúsculas, o es una mezcla (por ejemplo: "HoLa"). Sugerencia: apóyate en métodos de string como

### Tarea 8. 
Crea un programa que solicite al usuario:
- el tipo de cliente ("regular" o "vip"),
- el monto de compra,
- y si tiene cupón ("si"/"no").
  
Las reglas de beneficio son:
- Si el cliente es VIP y el monto ≥ 80, entonces se aplica un 20% descuento
- Si el cliente es VIP y el monto < 80, entonces se aplica un 10% descuento
- Si el cliente es regular y el monto ≥ 100 USD, entonces se aplica un 10% descuento
- Si tiene cupón y no aplicó descuento por reglas anteriores, entonces se aplica 5 USD de descuento fijo,
validando que el total luego del descuento no baje de 0 USD (no tiene sentido un total de -$1.50 por
ejemplo).
- Finalmente, el programa mostrará el total a pagar y explicar qué regla se aplicó (mensaje claro).


```python
def calcular_monto(tipo, monto, cupon):
    porcentaje = 0
    descuento = 0
    hay_descuento = False
    
    if tipo == 'vip' and monto >= 80 :
        porcentaje = 20
        descuento = (monto * porcentaje)/100
        hay_descuento = True
    elif tipo == 'vip' and monto > 0 and monto < 80 :
        porcentaje = 10
        descuento = (monto * porcentaje)/100
        hay_descuento = True
    elif tipo == 'regular' and monto >= 100 :
        porcentaje = 10
        descuento = (monto * porcentaje)/100
        hay_descuento = True
    if cupon == 'si' and not hay_descuento and monto > 5:
        descuento = 5
        
        
    return (descuento, porcentaje)
        
# Validar que digite datos correctos
client = input('Digite el tipo de cliente que es (regular/vip): ')
while (client.lower() != "regular" and client.lower() != "vip") :
    client = input('Digite el tipo de cliente nuevamente: (regular/vip)')

monto = input('Digite el monto: ')
while not monto.isdigit() :
    monto = input('Digite el monto correcto: ')
monto = int(monto)

has_cupon = input('Tiene cupón? (si/no): ')
while (has_cupon.lower() != "si" and has_cupon.lower() != "no") :
    has_cupon = input('Tiene cupón? (si/no)')

result = calcular_monto(client, monto, has_cupon)
print(f'El cliente es {client.upper()}, su monto es ${ monto - result[0]}, al aplicarle {result[1]}% al monto original ${monto}')
```

### Tarea 9. 
Solicita a, b, c (enteros > 0). Primero valida si pueden formar triángulo. Si es válido, clasifica:
- Equilátero, Isósceles o Escaleno
- Además, determina si es rectángulo usando Pitágoras.

Salida ejemplo:
- "Triángulo inválido"
- "Isósceles y rectángulo"
- "Escaleno (no rectángulo)"

```python
def validar_lado() :
    while True:
        try:
            num = int(input('Digite un lado: '))
            if num > 0:
                return num
            else:
                print('Debe ser mayor que 0')
        except ValueError:
            print('Debe ingresar un número válido')


def is_triangle_valid(num1, num2, num3) :
    first_sum = num1 + num2 
    second_sum = num1 + num3
    third_sum = num2 + num3
    if first_sum > num3 and second_sum > num2 and third_sum > num1 :
        return True
    else :
        return False


def type_triangle(num1, num2, num3) :
    lista = sorted([num1, num2, num3])
    pitagoras = (lista[0]**2 + lista[1]**2) == lista[2]**2
    
    if (num1 == num2 == num3):
        return 'Es un Triángulo Equilatero'
    elif (num1 == num2 or num1 == num3 or num2 == num3) :
        if (pitagoras) :
            return 'Es un Triángulo Isosceles, Rectangulo'
        return 'Es un Triángulo Isosceles'
    elif (pitagoras)  :
        return 'Es un Triángulo Escaleno, Rectángulo'
    else :
        return 'Es un Triángulo Escaleno'
    
a = validar_lado()
b = validar_lado()
c = validar_lado()

valid = is_triangle_valid(a, b, c)

if valid :
    result = type_triangle(a, b, c)
    print(result)
else :
    print('Triangulo Inválido')
```
