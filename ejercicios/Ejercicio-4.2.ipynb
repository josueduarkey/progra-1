# Ejercicio 1
Se desea calcular el tiempo de vuelo y la velocidad horizontal de un proyectil. Cada estudiante deberá
implementar funciones lambda para realizar los cálculos básicos a partir de los valores ingresados por
el usuario.

```python

import math

vel_inicial = float(input('Digite la velocidad inicial (m/s): '))
angulo = float(input('Digite el angulo de lanzamiento °: '))

# Convertir a radianes
angulo_rad = math.radians(angulo)

vel_horizontal = lambda vel, grado: round(vel * math.cos(angulo_rad), 2)
result_vel_horizontal = vel_horizontal(vel_inicial, angulo)
print(f'La velocidad horizontal es: {result_vel_horizontal} m/s')

vel_vertical = lambda vel, grado: round(vel * math.sin(angulo_rad), 2)
result_vel_vertical = vel_vertical(vel_inicial, angulo)
print(f'La velocidad vertical es: {result_vel_vertical} m/s')

tiempo_vuelo = lambda vel_y : (vel_y * 2)/9.8
result_tiempo_vuelo = tiempo_vuelo(result_vel_vertical)
print(f'El tiempo de vuelo es:  {round(result_tiempo_vuelo, 2)} s')
```

# Ejercicio 2


```python
import math

lista_angulos = [15, 30, 45, 60, 75]
v_inicial = 30
gravedad = 9.81

distancias = list(map(
    lambda ang: round((v_inicial**2 / gravedad) * math.sin(2 * math.radians(ang)), 2),
    lista_angulos
))

angulos_destacados = list(filter(
    lambda ang: round((v_inicial**2 / gravedad) * math.sin(2 * math.radians(ang)), 2) > 70,
    lista_angulos
))

print(f'Ángulos: {lista_angulos}')
print(f'Distancias calculadas: {distancias}')
print(f'Ángulos con alcance > 70 m: {angulos_destacados}')
```

```python
from functools import reduce
suma_total = reduce(lambda acc, x: acc + x, distancias)

promedio = suma_total / len(distancias)
maxima = reduce(lambda a, b: a if a > b else b, distancias)

sobre_promedio = [d for d in distancias if d > promedio]

print(f'Distancia promedio: {round(promedio, 2)} m')
print(f'Distancia máxima: {round(maxima, 2)} m')
print(f'Lanzamientos sobre el promedio: {sobre_promedio}')
```
