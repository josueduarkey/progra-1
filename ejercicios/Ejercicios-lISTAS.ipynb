```python
intensidades = []
for i in range(10):
    value = int(input(f'{i+1} - Ingresa los valores de la intensidad: '))s
    intensidades.append(value)

valor_max = max(intensidades)
valor_min = min(intensidades)

index_max = intensidades.index(valor_max)
index_min = intensidades.index(valor_min)

suma = sum(intensidades)
promedio = suma / 10

print(f'Intensidades: {intensidades}')
print(f'Máximo: {valor_max} (minuto {index_max + 1})')
print(f'Mínimo: {valor_min} (minuto {index_min + 1})')
print(f'Promedio: {promedio:.2f}}')
```

```python
anomalias = []
for value in intensidades:
    if value < 25 or value > 90:
        anomalias.append(value)
        
print(f'Intensidades: {intensidades}')
print(f'Anomalías: {anomalias}')
print(f'Mediciones normales: {len(intensidades) - len(anomalias)}')
print(f'Mediciones anómalas: {len(anomalias)}')
```

```python
normalizadas = []
if valor_max == valor_min:
    normalizadas = [0] * len(intensidades)
else :
    for value in intensidades:
        normal_value = (value - valor_min)/(valor_max - valor_min)
        correct_value = float(format(normal_value, ".2f"))
        normalizadas.append(correct_value)
    
print(f'Intensidades originales: {intensidades}')
print(f'Normalizadas: {normalizadas}')
```
