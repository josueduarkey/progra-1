### Tarea 1
Crear un programa que solicite al usuarios 7 números enteros. El programa mostrará cuantos de ellos son números pares


```python
cont_pares = 0
for i in range(1, 8):
    number = int(input(f'{i} - Digite un número: '))
    if number % 2 == 0:
        cont_pares += 1

print(f'Ingresaste {cont_pares} números pares')
```

### Tarea 2
Solicitar al usuario un número entero positivo N(asumir dato correcto). Luego mostrar en
pantallla la suma de todos los números multiplos de 5 desde 1 hasta N. 


```python
number = int(input('Digite un número: '))

suma = []
for i in range(1, number + 1):
    if i % 5 == 0:
        suma.append(i)
result = sum(suma)
print(f'La suma de todos los números múltiplos de 5 de 1 hasta {number} es {result}')
```
