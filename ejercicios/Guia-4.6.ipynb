```python
puntos = []
total_puntos = []
promedios = []
sumatoria_total = 0
for i in range(8):
    fila = input(f'Equipo {i+1} - puntos (separados por coma): ')
    equipo = list(map(int, fila.split(',')))
    puntos.append(equipo)
    
    total = sum(equipo)
    promedio = total / len(equipo)

    sumatoria_total += total
    total_puntos.append(total)
    promedios.append(promedio)

max_puntos = max(total_puntos)

lideres = []
for i in range(len(total_puntos)):
    if total_puntos[i] == max_puntos:
        lideres.append(i + 1)

for i in range(8):
    print(f"Equipo {i+1}: Total = {total_puntos[i]}, Promedio = {promedios[i]:.2f}")
    
print(f'El promedio de puntos total en el torneo es: {sumatoria_total/40}')
print(f"\nLíder: Equipo {lideres} con {max_puntos} puntos")
```
