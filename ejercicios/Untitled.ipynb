```python
from random import randint

numero = randint(1, 10)
cantidad_user = int(input('Digite la cantidad de intentos'))
win = False

for i in range(1, cant_user + 1):
    num_user = int(input('Adivine el número: '))
    if num_user == numero:
        print(f'Ha ganado el juego en el intento {i}')
        win = True
        break
    elif num_user > numero:
        print(f'Intento {i} - Es mayor tu número')
    else :
        print(f'Intento {i} - Es menor tu número')
if not win:
    print('Perdiste')
```
