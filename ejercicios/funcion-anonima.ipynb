```python
# Dado un conjunto de datos que representan temperaturas
# en grados celcius. Aplicar una función que convierta a grados a farenjwight a la lista usando map y lambda


Temperaturas = [0, 18, 23, 38, 15]

cel_to_far =list(map(lambda grados: grados * 1.8 + 32,Temperaturas))
print(f'Esta es la lista convertida: {cel_to_far}')
```
