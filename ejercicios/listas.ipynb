```python
# Crear dos listas vacías. Solicitando 10 números al usuario 
```

```python
pares = []
impares = []
for i in range(10):
    number = int(input('Digite un número: '))
    if number % 2 == 0:
        pares.append(number)
    else :
        impares.append(number)
print(f'Lista de números pares: {pares}')
print(f'Lista de números impares: {impares}')
```
