### 1. Introducción a las listas
Crea una lista y solicita en teclado las calificaciones de 8 estudiantes. Luego, identifica y muestra en
pantalla cuál es la primera nota, la última y la que está en la posición 3.

```python
grades = []

for i in range(8):
    grade = float(input('Digite una nota: '))
    grades.append(grade)
print(f'La primera nota es: {grades[0]}')
print(f'La última nota es: {grades[-1]}')
print(f'La 3era nota es: {grades[2]}')
```

### 2. Agregar y eliminar elementos
Actividad. Investiga como realizar las siguientes acciones:
- Agrega una nueva calificación al final de la lista.
- Inserta una calificación en la segunda posición.
- Elimina la primera calificación.
- Muestra la lista actualizada.

```python
last_grade = float(input('digite una nota: '))
grades.append(last_grade)
second_grade = float(input('digite la segunda nota: '))
grades.insert(1, second_grade)
grades.pop(0)
print(grades)
```

### 3. Recorrer listas
Recorre la lista y determina cuántos estudiantes aprobaron (nota mayor o igual a 6) y cuántos
reprobaron.


```python
aprobados = 0
reprobados = 0
for grade in grades:
    if grade >= 6.0:
        aprobados += 1
    else:
        reprobados += 1
print(f'Aprobaron: {aprobados} estudiantes')
print(f'Reprobaron: {reprobados} estudiantes')
```

### 4. Búsqueda y conteo
Pregunta al usuario una calificación. Indica si está en la lista y cuántas veces aparece. Si no aparece,
muestra un mensaje indicando que no se encontró.

```python
user_grade = float(input('Digite una nota a buscar: '))
cont = grades.count(user_grade)
if cont == 0:
    print('La nota no se encontró')
elif cont == 1:
    print('La nota existe y aparece 1 vez')
else:
    print(f'La nota existe y aparece {cont} veces')
    
```

### 5. Ordenamiento e inversión
Pre Ordena la lista de calificaciones de menor a mayor.
- Luego invierte el orden

```python
sort_list = sorted(grades)
print(f'Lista ordenada {sort_list}')

reves = sorted(sort_list, reverse=True)
print(f'Lista al reves {reves}')
```
