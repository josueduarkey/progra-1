```python
entrada = int(input('Digite la dimensión: '))
matriz = []
for i in range(entrada):
    fila = input(f'Fila {i+1}: ')
    fila = list(map(int, fila.split(',')))
    matriz.append(fila)
matriz_principal = []
matriz_secundaria = []
list_of_user_number = []
n = len(matriz)
numero_usuario = int(input('Digite el número que quiere encontrar: '))

for ri in range(len(matriz)):
    for ci in range(len(matriz[ri])):
        # Diagonal secundaria
        if matriz[ri][ci] == numero_usuario:
            list_of_user_number.append([ri,ci])
        if ri + ci == n - 1:
            matriz_secundaria.append(matriz[ri][ci])
        if ri == ci:
            matriz_principal.append(matriz[ri][ci])
print(matriz_principal)
print(matriz_secundaria)
```

```python
dimension = int(input('Digite la dimensión: '))
matriz = []
for i in range(dimension):
    fila = input(f'Fila {i+1}: ')
    fila = list(map(int, fila.split(',')))
    matriz.append(fila)


es_simetrica = True
    
for i in range(dimension):
    for j in range(dimension):
        if matriz[i][j] != matriz[j][i]:
            es_simetrica = False
            break
    if not es_simetrica:
        break

if es_simetrica:
    print("La matriz es simétrica")
else:
    print("La matriz no es simétrica")
```
