```python
class Book:
    def __init__(self, title, author, pag_num):
        self.title = title
        self.author = author
        self.pag_num = pag_num
        self.pag_read = 0

    def read(self, cantidad):
        self.pag_read += cantidad
        if self.pag_read > self.pag_num:
            self.pag_read = self.pag_num
            return
    def progreso(self):
        print(f'Has leido {self.pag_read} de {self.pag_num}')

    def terminado(self):
         return self.pag_read == self.pag_num

libro1 = Book('Cien años de soledad', 'Gabriel García Marquez', 500)
libro1.read(510)
libro1.progreso()
libro1.terminado()
```
